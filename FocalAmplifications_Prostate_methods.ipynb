# CNV Statistics of TCGA Prostate cancer samples and controls

A region is defined as gained when log2-ratio > 0.2<br>
Similarly, a region is defined as lost when log2-ratio < -0.2
<br><br>
A sample is considered to contain CNVs when either the cumulative size of amplifications or deletion is more than three standard deviations higher than that of control samples

```python
sample_count = 0
control_count = 0
amp_lengths_samples = list()
del_lengths_samples = list()
amp_lengths_controls = list()
del_lengths_controls = list()
            
for control_sample in controls.values():
   cum_size_amp = 0
   cum_size_del = 0
   control_count += 1
   for cnv in control_sample.CNV:
        if cnv.amplified:
            cum_size_amp += cnv.size
        if cnv.deleted:
            cum_size_del += cnv.size
   amp_lengths_controls.append(cum_size_amp)
   del_lengths_controls.append(cum_size_del)
            

controls_amp_lengths_series=pandas.Series(amp_lengths_controls)  
controls_del_lengths_series=pandas.Series(del_lengths_controls)   
mean_amp_length_controls = controls_amp_lengths_series.mean() 
mean_del_length_controls = controls_del_lengths_series.mean()
std_amp_length_controls = controls_amp_lengths_series.std() 
std_del_length_controls = controls_del_lengths_series.std()

samples_amplified = 0
samples_deleted = 0
samples_both = 0
samples_either = 0

for sample in samples.values():
   cum_size_amp = 0
   cum_size_del = 0
   sample_count += 1
   for cnv in sample.CNV:
        if cnv.amplified:
            cum_size_amp += cnv.size
        if cnv.deleted:
            cum_size_del += cnv.size
   amp_z_score = (cum_size_amp - mean_amp_length_controls) / std_amp_length_controls
   del_z_score = (cum_size_del - mean_del_length_controls) / std_del_length_controls
   if ( amp_z_score > 3):
       samples_amplified += 1
   if ( del_z_score > 3):
       samples_deleted += 1
   if ( amp_z_score > 3 and del_z_score > 3):
       samples_both += 1
   if ( amp_z_score > 3 or del_z_score > 3):
       samples_either += 1
   amp_lengths_samples.append(cum_size_amp)
   del_lengths_samples.append(cum_size_del)
   
samples_amp_lengths_series=pandas.Series(amp_lengths_samples)  
samples_del_lengths_series=pandas.Series(del_lengths_samples)  

TCGA_samples_amp_lengths = samples_amp_lengths_series
TCGA_samples_del_lengths = samples_del_lengths_series


print "TCGA -samples"
print "-------------"
print ""
print "Samples: "+str(sample_count)
print "Samples with amplifications: "+str(samples_amplified)
print "Samples with deletions: "+str(samples_deleted)
print "Samples with deletions and amplifications: "+str(samples_both)
print "Samples with deletions or amplifications: "+str(samples_either)
print "Controls: "+str(control_count)
print "----------------"
print "Mean length amp samples: "+str(samples_amp_lengths_series.mean())+"    % of genome: "+str(samples_amp_lengths_series.mean()/30956774.12) 
print "Mean length del samples: "+str(samples_del_lengths_series.mean())+"    % of genome: "+str(samples_del_lengths_series.mean()/30956774.12) 
print "Mean length amp controls: "+str(mean_amp_length_controls)+"    % of genome: "+str(mean_amp_length_controls/30956774.12) 
print "Mean length del controls: "+str(mean_del_length_controls)+"    % of genome: "+str(mean_del_length_controls/30956774.12) 
print "Std length amp controls: "+str(std_amp_length_controls)
print "Std length del controls: "+str(std_del_length_controls)
```

```python
sample_count = 0
control_count = 0
amp_lengths_samples = list()
del_lengths_samples = list()
amp_lengths_controls = list()
del_lengths_controls = list()
centromers = {"chr1":[121500000,128900000],"chr2":[90500000,96800000],"chr3":[87900000,91000000],"chr4":[48200000,52700000],"chr5":[46100000,48400000],"chr6":[58700000,63300000],"chr7":[58000000,61700000],"chr8":[43100000,48100000],"chr9":[47300000,50700000],"chr10":[38000000,42300000],"chr11":[51600000,55700000],"chr12":[33300000,38200000],"chr13":[1,19500000],"chr14":[1,17600000],"chr15":[1,19000000],"chr16":[34600000,38600000],"chr17":[22200000,25800000],"chr18":[15400000,19000000],"chr19":[24400000,28600000],"chr20":[25600000,29400000],"chr21":[1,14300000],"chr22":[1,17900000],"chrX":[60600000,63000000],"chrY":[11600000,12500000]}

def substract_centromer(chrom,start,end):
    #return total size of not overlapping the centromer
    if end < centromers[chrom][0] or start > centromers[chrom][0]:
        return (end-start)
    
    #return size 0 when segment is totally within centromer
    elif start > centromers[chrom][0] and end < centromers[chrom][1]:
        return 0
    
    #return nonoverlapping size when partially overlapping a centromere
    elif start > centromers[chrom][0] and start < centromers[chrom][1] and end > centromers[chrom][1]:
        return (end - centromers[chrom][1])
    elif end < centromers[chrom][1] and end > centromers[chrom][0] and start < centromers[chrom][0]:
        return (centromers[chrom][0] - start)
    
    #return segment size minus centromere size when segment spans a acentromere
    elif start < centromers[chrom][0] and end > centromers[chrom][1]:
        return (end - start - (centromers[chrom][1]-centromers[chrom][0]))
            
for control_sample in controls.values():
   cum_size_amp = 0
   cum_size_del = 0
   control_count += 1
   for cnv in control_sample.CNV:
        if cnv.amplified:
            cum_size_amp +=  substract_centromer(cnv.chrom,cnv.start,cnv.end)
        if cnv.deleted:
            cum_size_del +=  substract_centromer(cnv.chrom,cnv.start,cnv.end)
   amp_lengths_controls.append(cum_size_amp)
   del_lengths_controls.append(cum_size_del)
            

controls_amp_lengths_series=pandas.Series(amp_lengths_controls)  
controls_del_lengths_series=pandas.Series(del_lengths_controls)   
mean_amp_length_controls = controls_amp_lengths_series.mean() 
mean_del_length_controls = controls_del_lengths_series.mean()
std_amp_length_controls = controls_amp_lengths_series.std() 
std_del_length_controls = controls_del_lengths_series.std()

samples_amplified = 0
samples_deleted = 0
samples_both = 0
samples_either = 0

for sample in samples.values():
   cum_size_amp = 0
   cum_size_del = 0
   sample_count += 1
   for cnv in sample.CNV:
        if cnv.amplified:
            cum_size_amp +=  substract_centromer(cnv.chrom,cnv.start,cnv.end)
        if cnv.deleted:
            cum_size_del +=  substract_centromer(cnv.chrom,cnv.start,cnv.end)
   amp_z_score = (cum_size_amp - mean_amp_length_controls) / std_amp_length_controls
   del_z_score = (cum_size_del - mean_del_length_controls) / std_del_length_controls
   if ( amp_z_score > 3):
       samples_amplified += 1
   if ( del_z_score > 3):
       samples_deleted += 1
   if ( amp_z_score > 3 and del_z_score > 3):
       samples_both += 1
   if ( amp_z_score > 3 or del_z_score > 3):
       samples_either += 1
   amp_lengths_samples.append(cum_size_amp)
   del_lengths_samples.append(cum_size_del)
   
samples_amp_lengths_series=pandas.Series(amp_lengths_samples)  
samples_del_lengths_series=pandas.Series(del_lengths_samples)  

TCGA_samples_amp_lengths = samples_amp_lengths_series
TCGA_samples_del_lengths = samples_del_lengths_series


print "TCGA -samples (centromer removed)"
print "-------------"
print ""
print "Samples: "+str(sample_count)
print "Samples with amplifications: "+str(samples_amplified)
print "Samples with deletions: "+str(samples_deleted)
print "Samples with deletions and amplifications: "+str(samples_both)
print "Samples with deletions or amplifications: "+str(samples_either)
print "Controls: "+str(control_count)
print "----------------"
print "Mean length amp samples (centromer removed): "+str(samples_amp_lengths_series.mean())+"    % of genome: "+str(samples_amp_lengths_series.mean()/30956774.12) 
print "Mean length del samples (centromer removed): "+str(samples_del_lengths_series.mean())+"    % of genome: "+str(samples_del_lengths_series.mean()/30956774.12) 
print "Mean length amp controls (centromer removed): "+str(mean_amp_length_controls)+"    % of genome: "+str(mean_amp_length_controls/30956774.12) 
print "Mean length del controls (centromer removed): "+str(mean_del_length_controls)+"    % of genome: "+str(mean_del_length_controls/30956774.12) 
print "Std length amp controls: "+str(std_amp_length_controls)
print "Std length del controls: "+str(std_del_length_controls)
```
